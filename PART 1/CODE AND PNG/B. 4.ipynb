```python
# Scatter Plot: Gold vs Silver Annual Average Prices

import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display
import os

# Load datasets
gold = pd.read_csv(r"E:\UE class schedule\DATA SCIENCE\DSB Project\Gold 100years.csv")
silver = pd.read_csv(r"E:\UE class schedule\DATA SCIENCE\DSB Project\silver 100 years.csv")

# Convert Date columns
gold["Date"] = pd.to_datetime(gold["Date"], dayfirst=True)
silver["Date"] = pd.to_datetime(silver["Date"], dayfirst=True)

# Extract Year
gold["Year"] = gold["Date"].dt.year
silver["Year"] = silver["Date"].dt.year

# Calculate annual average prices
gold_yearly = gold.groupby("Year")["Value"].mean().reset_index()
silver_yearly = silver.groupby("Year")["Value"].mean().reset_index()

# Merge datasets by Year
merged = pd.merge(
    gold_yearly,
    silver_yearly,
    on="Year",
    suffixes=("_Gold", "_Silver")
)

# Display first few rows
print("Annual Average Gold and Silver Prices")
display(merged.head())

# Create scatter plot
plt.figure(figsize=(8,6))
plt.scatter(
    merged["Value_Gold"],
    merged["Value_Silver"]
)

plt.title("Gold vs Silver Annual Average Prices")
plt.xlabel("Average Gold Price")
plt.ylabel("Average Silver Price")
plt.grid(True)

plt.show()

# Save figure
save_folder = r"E:\UE class schedule\DATA SCIENCE\DSB Project\Figures"
os.makedirs(save_folder, exist_ok=True)

plt.figure(figsize=(8,6))
plt.scatter(
    merged["Value_Gold"],
    merged["Value_Silver"]
)

plt.title("Gold vs Silver Annual Average Prices")
plt.xlabel("Average Gold Price")
plt.ylabel("Average Silver Price")
plt.grid(True)

plt.savefig(
    os.path.join(save_folder, "Q20_Gold_vs_Silver_Scatter_Plot.png"),
    bbox_inches="tight",
    dpi=300
)

plt.close()

print("Scatter plot saved successfully!")
```
