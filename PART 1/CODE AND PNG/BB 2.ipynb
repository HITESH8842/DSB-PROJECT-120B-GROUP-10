```python
# Correlation Table: Gold, Silver and Brent Oil

import pandas as pd
import os
import matplotlib.pyplot as plt
from IPython.display import display

# Load datasets
gold = pd.read_csv(r"E:\UE class schedule\DATA SCIENCE\DSB Project\Gold 100years.csv")
silver = pd.read_csv(r"E:\UE class schedule\DATA SCIENCE\DSB Project\silver 100 years.csv")
brent = pd.read_csv(r"E:\UE class schedule\DATA SCIENCE\DSB Project\Brent Oil.csv")

# Convert Date columns
gold["Date"] = pd.to_datetime(gold["Date"], dayfirst=True)
silver["Date"] = pd.to_datetime(silver["Date"], dayfirst=True)
brent["Date"] = pd.to_datetime(brent["Date"], dayfirst=True)

# Extract Year
gold["Year"] = gold["Date"].dt.year
silver["Year"] = silver["Date"].dt.year
brent["Year"] = brent["Date"].dt.year

# Annual average prices
gold_yearly = gold.groupby("Year")["Value"].mean().reset_index()
silver_yearly = silver.groupby("Year")["Value"].mean().reset_index()
brent_yearly = brent.groupby("Year")["Value"].mean().reset_index()

# Merge datasets
merged = gold_yearly.merge(
    silver_yearly,
    on="Year",
    suffixes=("_Gold", "_Silver")
)

merged = merged.merge(
    brent_yearly,
    on="Year"
)

merged.rename(
    columns={"Value": "Value_Brent"},
    inplace=True
)

# Correlation table
correlation_table = merged[
    ["Value_Gold", "Value_Silver", "Value_Brent"]
].corr()

# Display in Jupyter
print("Correlation Table")
display(correlation_table)

# Create folder
save_folder = r"E:\UE class schedule\DATA SCIENCE\DSB Project\Tables"
os.makedirs(save_folder, exist_ok=True)

# Save CSV
correlation_table.to_csv(
    os.path.join(
        save_folder,
        "QB2_Correlation_Table_Gold_Silver_Brent.csv"
    )
)

# Save PNG
fig, ax = plt.subplots(figsize=(6,4))
ax.axis('off')

table = ax.table(
    cellText=correlation_table.round(3).values,
    rowLabels=correlation_table.index,
    colLabels=correlation_table.columns,
    cellLoc='center',
    loc='center'
)

table.auto_set_font_size(False)
table.set_fontsize(10)
table.scale(1.2, 1.5)

plt.savefig(
    os.path.join(
        save_folder,
        "QB2_Correlation_Table_Gold_Silver_Brent.png"
    ),
    bbox_inches='tight',
    dpi=300
)

plt.close()

print("QB2 table saved successfully!")
```
